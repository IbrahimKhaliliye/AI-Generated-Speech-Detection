```python
import numpy as np
import librosa
import pandas as pd
import librosa
import numpy as np

def extract_features1(wav_file, sr=22050):
    # Load the audio file
    y, sr = librosa.load(wav_file, sr=sr)

    # Extract features
    chroma_stft = librosa.feature.chroma_stft(y=y, sr=sr)
    rms = librosa.feature.rms(y=y)
    spectral_centroid = librosa.feature.spectral_centroid(y=y, sr=sr)
    spectral_bandwidth = librosa.feature.spectral_bandwidth(y=y, sr=sr)
    rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
    zero_crossing_rate = librosa.feature.zero_crossing_rate(y)

    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20)

    # Aggregate features
    features = [np.mean(chroma_stft), np.mean(rms), np.mean(spectral_centroid),
                np.mean(spectral_bandwidth), np.mean(rolloff), np.mean(zero_crossing_rate)]
    for mfcc in mfccs:
        features.append(np.mean(mfcc))

    return features
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

def plot_mel_spectrogram(wav_file_path, n_mels=128, fmax=8000):
    # Load the WAV file
    y, sr = librosa.load(wav_file_path, sr=None)

    # Compute the Mel spectrogram
    S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=n_mels, fmax=fmax)

    # Convert the power spectrogram (amplitude squared) to decibel (dB) units
    S_dB = librosa.power_to_db(S, ref=np.max)

    # Plot the Mel spectrogram
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel', fmax=fmax)
    plt.colorbar(format='%+2.0f dB')
    plt.title('Mel-frequency spectrogram')
    plt.tight_layout()
    plt.show()

    return S_dB

```

```python
import librosa
import matplotlib.pyplot as plt

def plot_waveform(wav_file_path):
    # Load the WAV file
    y, sr = librosa.load(wav_file_path, sr=None)

    # Plot the waveform
    plt.figure(figsize=(10, 4))
    librosa.display.waveshow(y, sr=sr)
    plt.title('Waveform')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.tight_layout()
    plt.show()
```

```python
import audioread
def preprocess_audio(audio_path):

  # Load audio using librosa
  try:
    y, sr = librosa.load(audio_path, sr=None)
  except librosa.util.exceptions.ParameterError as e:
    print(f"Error loading {audio_path} with librosa: {e}")
    return None
  except audioread.NoBackendError:
    print(f"Error loading {audio_path}: No backend available for audioread.")
    return None
  except Exception as e:
    print(f"Unexpected error loading {audio_path}: {e}")
    return None

  y, sr = librosa.load(audio_path)
  # Resample to a consistent sampling rate (e.g., 22050 Hz)
  if sr != target_sr:
    y = librosa.effects.resample(y, sr, target_sr)
  # Extract Mel spectrogram features (customizable parameters)
  mel_spectrogram = librosa.feature.melspectrogram(y=y, n_mels=128, sr=target_sr)
  # Apply normalization (e.g., scaling to 0-1 range)
  normalized_mel = librosa.power_to_db(mel_spectrogram).astype("float32") / 80
  mel_spectrogram = mel_spectrogram.astype('float32')
  mel_spect_db = librosa.power_to_db(mel_spectrogram, ref=np.max)
  return mel_spect_db

# Define target sampling rate
target_sr = 22050
```

```python
def segment_mel_spectrogram(mel_spect, segment_length=128, hop_length=64):
    segments = []
    num_segments = (mel_spect.shape[1] - segment_length) // hop_length + 1
    for i in range(num_segments):
        start = i * hop_length
        end = start + segment_length
        segment = mel_spect[:, start:end]
        segments.append(segment)
    return segments

```
